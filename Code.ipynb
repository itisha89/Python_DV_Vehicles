### Data cleaning

```python
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.4f' % x)

# Exclude the first two columns
df_without_first_two_columns = df.iloc[:, 2:]

# Create a new DataFrame to store the cleaned data
cleaned_df = df_without_first_two_columns.copy()

# Identifying NULL or Missing or NA values
print('\n\nNull/Missing/NA Values Check:\n')
print(cleaned_df.isnull().sum())

# Identify rows with missing values
rows_with_empty_values = cleaned_df[cleaned_df.isnull().sum(axis=1) > 0]
print('Row indices having NA/null values: \n', rows_with_empty_values.index.tolist())

# Replace the missing values in each row with the mean value of that row in the new DataFrame
for row_index, row in rows_with_empty_values.iterrows():
    mean_value = row.mean()
    cleaned_df.loc[row_index] = row.fillna(mean_value)

# Check NULL/Missing/NA Values After Cleaning
print('\n\nCheck Null/Missing/NA Values after cleaning:\n')
print(cleaned_df.isnull().sum())

# Print the Cleaned DataFrame
print('\n\nPrinting cleaned data frame:\n')
print(cleaned_df.tail(5))

# Checking for duplicates in the cleaned DataFrame
print('\n\nDuplicate Data Check in cleaned DataFrame:\n')
print('Count of duplicates in cleaned data: ', cleaned_df.duplicated().sum(), '\n')

# Creating the final cleaned DataFrame with all the columns present
cleaned_df2= pd.DataFrame()
cleaned_df2['Category'] = df['Category']
cleaned_df2['Segment'] = df['Segment']
merged_cleaned_df = pd.concat([cleaned_df2, cleaned_df], axis=1)
merged_cleaned_df
```

### Compare the production of passenger vehicles and commercial vehicles in India from 2001-02 to 2012-13, in a single visual.
    

```python
selected_columns_rows = merged_cleaned_df.loc[[2, 9], ["2001-02", "2002-03", "2003-04", 
                                     "2004-05", "2005-06", "2006-07", 
                                     "2007-08","2008-09", "2009-10", 
                                     "2010-11", "2011-12", "2012-13"]]

# Calculate the width of each bar
bar_width = 0.40

# Create an array of y positions for each category
y = np.arange(len(selected_columns_rows.columns))

# Create a vertical bar graph with two sets of values
plt.barh(y - bar_width/2, selected_columns_rows.iloc[0], bar_width, label='Passenger Vehicles')
plt.barh(y + bar_width/2, selected_columns_rows.iloc[1], bar_width, label='Commercial Vehicles')

# Adding labels, title, and legend
plt.xlabel('Vehicles Production in millions')
plt.ylabel('Year')
plt.title('Passenger Vehicles and Commercial Vehicles')
plt.yticks(y, selected_columns_rows.columns)
plt.legend()

# Display the plot
plt.show()
```

2. Justifying the choice of visual
A--> We chose bar graph because of the following reasons
       
       a. It is best to show year-on-year trends
       
       b. It is good for comparative analysis
       
       c. It is flexible, readable, and has visual emphasis on the values 
       
3. Pre-attentive attributes used
A--> Following attributes are used
       
       a. Color: Different colors are used for bars representing passenger vehicles and commercial vehicles.
       
       b. Length: The length of the bars is a pre-attentive attribute that directly represents the quantity being visualized. Longer bars indicate larger values, making it easy to compare quantities.
       
       c. Position: The position of the bars along the common axis (usually the x-axis) is crucial for comparing data points.
       
       The distance between bars and their positions along the axis are pre-attentive cues for understanding relative values.

4. Gestalt principles used
A--> Following gestalt principles are used.
       a. Proximity: The Gestalt principle of proximity suggests that objects that are close to each other are perceived as related or part of the same group. In a bar graph, bars that are close to each other on the same axis are perceived as related data points.
       
       b. Similarity: The principle of similarity states that objects that share similar attributes, such as color or shape, are perceived as belonging to the same group. Similar colors or styles are often used to group bars representing related data.
       
       c. Figure-Ground: Figure-ground perception helps distinguish between the main data (the "figure") and the background. In a bar graph, the bars themselves represent the figure, while the background is kept as white.

### Compare the production of 2 wheeler and three wheelers in India from 2001-02 to 2012-13, in a single visual.

    b. The growth in the production of 2 -wheelers has been gradual and exponential during the year 2007-08 onwards
    
    c. The production of 3-wheelers has been less as compared to that of 2-wheelers
   
    d. The growth in the production of 3-wheelers has been steady and modest.

2. Justifying the choice of visual
A--> A stacked area chart allows you to represent the total production while showing the contributions of both two-wheelers and three-wheelers as separate segments over time. This enables viewers to see not only the overall trend in production but also the relative proportions of each vehicle type.

3. Pre-attentive attributes used

    b. Area: The area enclosed by each segment of the stacked chart is a pre-attentive attribute. Larger areas represent greater quantities of production, allowing viewers to quickly assess the relative production levels of Three Wheelers and Two Wheelers.

4. Gestalt principles used
A--> a.  Proximity: The principle of proximity suggests that elements that are close to each other are perceived as related. In the stacked area chart, segments within the same year (aligned horizontally) are perceived as related data points.
     
     b. Similarity: The principle of similarity applies to the use of similar colors to represent related data. Segments with the same color are perceived as part of the same category or group, reinforcing the idea of similarity.
     
     c. Figure-Ground: Effective use of figure-ground perception distinguishes the primary data (the stacked area segments) from the background, enhancing clarity and focusing attention on the data.

```python
categories = selected_columns_rows_categories.columns

values = []

for column_name in categories:
    values.append(selected_columns_rows_categories.get(column_name))


values = np.array(values)
total_values = np.sum(values, axis=1)
percentages = (values.T / total_values * 100).T

plt.figure(figsize=(10, 60))

for i, category in enumerate(categories):
    plt.subplot(len(categories), 1, i + 1)
    plt.pie(percentages[i], labels=['Passenger Vehicles', 'Commercial Vehicles', 'Three Wheelers', 'Two Wheelers'], autopct='%1.1f%%', startangle=140)
    plt.title(category+' Year')
    print("\n")

plt.tight_layout()
plt.show()
```

2. Justifying the choice of visual

A--> a.  Showing Parts of a Whole: Pie charts are excellent for displaying how individual parts (categories) contribute to a whole (the total). They provide a clear representation of proportions and percentages.

     b. Simplicity and Clarity: Pie charts are simple and easy to understand. They convey information in a visually straightforward manner, making them accessible to a wide audience.
     
     c. Limited Categories: When you have a small number of categories or segments to represent, pie charts can be effective. 
     
     d. Highlighting Dominant Categories: Pie charts emphasize the largest categories by positioning them near the top, making it easy to identify which segments are most significant.

3. Pre-attentive attributes used
A--> a. Color: Color can be used to distinguish between segments in a pie chart. Each segment can have a different color to aid in differentiation.

4. Gestalt principles used

A--> a. Closure: The principle of closure is evident in a pie chart as viewers perceive the entire circle as a whole. The circle appears complete, and the individual segments visually "close" the circle.

     b. Proximity: Proximity is relevant in pie charts when segments are close to each other. Segments that are adjacent in the circle are perceived as related and part of the same chart.
